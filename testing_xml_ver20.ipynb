## Cleaning Text

```python
def clean_text(text):

    text = text.replace('\r', '\n')
    text = text.replace(u'\xa0', ' ')
    text = text.replace(u'\x96', '-')
    text = text.replace(u'\xad', '-')
    text = text.replace('â', '')
    # text = text.replace(u'\u2014', 'â')
    # text = text.replace(u'\u2013', 'â')
    return text
```
